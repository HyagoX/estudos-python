```python
# Filtra as linhas em que o valor na coluna 'Produto' Seja 'Cabo HDMI e que o valor na coluna 'Cliente' seja igual a 'Cliente_24809'
df_vendas[(df_vendas['Produto'] == 'Cabo HDMI') & (df_vendas['Cliente'] == 'Cliente_24809')]
```

```python
# Filtra as linhas em que o valor na coluna 'Produto' Seja 'Cabo HDMI OU que o valor na coluna 'Cliente' seja igual a 'Cliente_24809'
df_vendas[(df_vendas['Produto'] == 'Cabo HDMI') | (df_vendas['Cliente'] == 'Cliente_24809')]
```

```python
# Mostra somente as linhas em que o valor da coluna Qtd seja maior que 2, tambem podemos usar menos que, diferente de, etc
df_vendas[df_vendas['Qtd'] > 2]
```

```python
# Pega todas as linhas em que o código do cliente termina com o número 6, tambem temos as funções startwith()
df_vendas[df_vendas['Cliente'].str.endswith('6')]
```

```python
# Filtra todas as linhas que fazem parte das lojas de Belo Horizonte ou do Rio de Janeiro
df_vendas[df_vendas['Loja'].isin(['Belo Horizonte', 'Rio de Janeiro'])]
```

```python
# Aqui fazemos o contrario do bloco anterior, estamos verificando somente as linhas de lojas que NÃO sao de Belo Horizonte ou do Rio de Janeiro
df_vendas[~df_vendas['Loja'].isin(['Belo Horizonte', 'Rio de Janeiro'])]
```

```python
# LOC = Locate Based on Label. Esse código localiza as linhas em que o cliente tem o código Cliente_3274
df_vendas.loc[df_vendas['Cliente'] == 'Cliente_3274']
```

```python
# Este código, primeiro localiza a linha especifica com o nome do Cliente, e troca pelo seu Nome dele
df_vendas.loc[df_vendas['Cliente'] == 'Cliente_3274', 'Cliente'] = 'Lucio'
```

```python
df_vendas.loc[df_vendas['Cliente'] == 'Lucio']
```

```python

# Transforma os textos com letras todas em maiusculo ou todas em minuculo para somente a primeira letra em maiuscula
# cidade = 'Belo Horizonte'
# df_vendas.loc[(df_vendas['Loja'] == cidade.lower()) | (df_vendas['Loja'] == cidade.upper()), 'Loja'] = cidade

# Uma forma mais simples e melhor de fazer isso, é usando o comando .str.title(), que já faz isso manualmente
df_vendas['Loja'] = df_vendas['Loja'].str.title()
```
